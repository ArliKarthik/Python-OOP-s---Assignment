#Python OOPS

1. What is Object-Oriented Programming (OOP)?
   - Programming paradigm where data and functions are bundled into objects, which are instances of classes.
   - Helps in code reusability, modularity, and easier maintenance through concepts like encapsulation, inheritance, and polymorphism.

2. What is a class in OOP?
   - A blueprint or template for creating objects; defines attributes (data) and methods (functions).
   - Objects created from a class share the structure and behavior defined in the class.

3. What is an object in OOP?
   - An instance of a class that contains actual values for the attributes defined in the class.
   - Can access the class’s methods and perform actions defined in the class.

4. What is the difference between abstraction and encapsulation?
   - Abstraction: Hides implementation details and shows only the essential features to the user.
     - Focuses on what an object does, not how it does it.
   - Encapsulation: Hides internal data of an object using private variables and provides access via methods.
     - Focuses on how data is protected and controlled within the object.

5. What are dunder methods in Python?
   - Special methods with double underscores before and after the name (e.g., __init__, __str__) used to define object behavior.
   - Allow custom behavior for built-in operations like addition, string conversion, and iteration.

6. Explain the concept of inheritance in OOP.
   - Allows a class (child/subclass) to reuse attributes and methods of another class (parent/superclass).
   - Promotes code reusability and makes it easier to create hierarchies of related classes.

7. What is polymorphism in OOP?
   - Allows objects of different classes to be accessed through the same interface, enabling the same operation to behave differently.
   - Supports method overriding (same method name, different behavior) and operator overloading.

8. How is encapsulation achieved in Python.
   - Achieved by making variables private using a double underscore prefix (__variable) to restrict direct access.
   - Access to private data is provided through getter and setter methods, controlling how data is read or modified.

9. What is a constructor in Python?
   - A special method __init__() used to initialize an object’s attributes when it is created.
   - Automatically called when an object is instantiated, allowing you to set default or initial values.

10. What are class and static methods in Python?
    - Class Method: Defined using the @classmethod decorator and takes cls as the first parameter.
      - Can access or modify class-level attributes, not instance-level attributes.
    - Static Method: Defined using the @staticmethod decorator and does not take self or cls.
      - Cannot access instance or class attributes directly; used for utility functions related to the class.

11. What is method overloading in Python?
    - Refers to defining multiple methods with the same name but different number or type of parameters in the same class.
    - Python does not support traditional method overloading, but it can be simulated using default arguments or *args/**kwargs.

12. What is method overriding in OOP?
    - Occurs when a child class provides its own implementation of a method that is already defined in the parent class.
    - Allows a subclass to change or extend the behavior of the inherited method.

13. What is a property decorator in Python?
    - @property allows a method to be accessed like an attribute, providing controlled access to private variables.
    - Helps implement getter, setter, and deleter methods in a clean and readable way.

14. Why is polymorphism important in OOP?
    - Allows different objects to be accessed through the same interface, enabling code to be more flexible and extensible.
    - Supports code reusability and simplifies maintenance by allowing method overriding and operator overloading.

15. What is an abstract class in Python?
    - A class that cannot be instantiated and is designed to be a base class for other classes.
    - Can contain abstract methods (declared but not implemented) using the @abstractmethod decorator from the abc module.

16. What are the advantages of OOP?
    - Code Reusability & Modularity: Classes and objects allow code to be reused, and large programs can be organized into smaller, manageable modules.
    - Easier Maintenance & Extensibility: Inheritance, encapsulation, and polymorphism make it easy to update, extend, and debug code without affecting other parts.

17. What is the difference between a class variable and an instance variable?
    - Class variable: Shared by all instances of the class; defined inside the class but outside any method.
      - Changes to the variable affect all objects of that class.
    - Instance variable: Unique to each object; defined using self inside methods like __init__().
      - Changes affect only that particular object.

18. What is multiple inheritance in Python?
    - A class can inherit from more than one parent class, gaining attributes and methods from all of them.
    - Enables code reuse and combining behaviors from multiple classes, but can lead to complexity (e.g., method resolution order).

19. Explain the purpose of "__str__" and "__repr__" methods in Python.
    - __str__: Returns a readable string representation of an object, meant for end-users.
      - Called by print() or str() on an object.
    - __repr__: Returns an unambiguous string representation of an object, meant for developers/debugging.
      - Called by repr() and used in interactive sessions.

20. What is the significance of the ‘super()’ function in Python?
    - Allows a child class to call methods or constructors of its parent class without explicitly naming it.
    - Useful for code reuse and handling multiple inheritance, ensuring proper method resolution order (MRO).

21. What is the significance of the __del__ method in Python?
    - Called when an object is about to be destroyed (garbage collected), allowing cleanup of resources like files or network connections.
    - Helps perform finalization tasks automatically before the object is removed from memory.

22. What is the difference between @staticmethod and @classmethod in Python?
    - @staticmethod: Does not take self or cls as the first parameter; cannot access instance or class variables directly.
      - Used for utility functions related to the class but independent of class/instance state.
    - @classmethod: Takes cls as the first parameter; can access or modify class variables.
      - Used for methods that affect the class as a whole rather than individual instances.

23. How does polymorphism work in Python with inheritance?
    - A child class can override a method of its parent class, allowing the same method call to behave differently depending on the object.
    - Enables uniform interface: code can call the same method on different objects without worrying about their specific types.

24. What is method chaining in Python OOP?
    - Technique where multiple methods are called in a single statement, with each method returning the object itself.
    - Makes code concise and readable, especially for performing multiple operations on the same object.

25. What is the purpose of the __call__ method in Python?
    - Allows an object of a class to be called like a regular function using parentheses ().
    - Enables objects to have function-like behavior, useful for callbacks or custom callable objects.

```python
#1. Create a parent class Animal with a method speak() that prints a generic message. Create a child class Dog that overrides the speak() method to print "Bark!".
class Animal:
    def speak(self):
        print("This animal makes a sound")

class Dog(Animal):
    def speak(self):
        print("Bark!")

a = Animal()
a.speak()

d = Dog()
d.speak()
```

```python
#2. Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle from it and implement the area() method in both.
from abc import ABC, abstractmethod
import math

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

c = Circle(5)
r = Rectangle(4, 6)
print("Circle area:", c.area())
print("Rectangle area:", r.area())
```

```python
#3.  Implement a multi-level inheritance scenario where a class Vehicle has an attribute type. Derive a class Car and further derive a class ElectricCar that adds a battery attribute.
class Vehicle:
    def __init__(self, type):
        self.type = type

class Car(Vehicle):
    def __init__(self, type, brand):
        super().__init__(type)
        self.brand = brand

class ElectricCar(Car):
    def __init__(self, type, brand, battery):
        super().__init__(type, brand)
        self.battery = battery

    def show_info(self):
        print(f"Type: {self.type}, Brand: {self.brand}, Battery: {self.battery} kWh")

ecar = ElectricCar("Car", "Tesla", 100)
ecar.show_info()
```

```python
#4. Demonstrate polymorphism by creating a base class Bird with a method fly(). Create two derived classes Sparrow and Penguin that override the fly() method.
class Bird:
    def fly(self):
        print("Some birds can fly")

class Sparrow(Bird):
    def fly(self):
        print("Sparrow can fly")

class Penguin(Bird):
    def fly(self):
        print("Penguin cannot fly")

birds = [Sparrow(), Penguin()]
for bird in birds:
    bird.fly()




```

```python
#5. Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes balance and methods to deposit, withdraw, and check balance.
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Insufficient balance")

    def check_balance(self):
        return self.__balance

account = BankAccount(1000)
account.deposit(500)
account.withdraw(300)
print(account.check_balance())

```

```python
#7. Create a class MathOperations with a class method add_numbers() to add two numbers and a static method subtract_numbers() to subtract two numbers.
class MathOperations:
    @classmethod
    def add_numbers(cls, a, b):
        return a + b

    @staticmethod
    def subtract_numbers(a, b):
        return a - b

x = MathOperations.add_numbers(5, 3)
y = MathOperations.subtract_numbers(10, 4)
print(x)
print(y)
```

```python
#9. Write a class Fraction with attributes numerator and denominator. Override the str method to display the fraction as "numerator/denominator".
class Fraction:
    def __init__(self, numerator, denominator):
        self.numerator = numerator
        self.denominator = denominator

    def __str__(self):
        return f"{self.numerator}/{self.denominator}"

num = Fraction(3, 4)
print(num)
```

```python
#10.  Demonstrate operator overloading by creating a class Vector and overriding the add method to add two vectors.
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

v1 = Vector(1, 2)
v2 = Vector(3, 4)
v3 = v1 + v2
print(v3.x, v3.y)
```

```python
#12.  Implement a class Student with attributes name and grades. Create a method average_grade() to compute the average of the grades.
class Student:
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades

    def average_grade(self):
        return sum(self.grades) / len(self.grades)

candidate = Student("John", [85, 90, 78, 92])
print(candidate.average_grade())
```

```python
#14. Create a class Employee with a method calculate_salary() that computes the salary based on hours worked and hourly rate. Create a derived class Manager that adds a bonus to the salary.
class Employee:
    def __init__(self, name, hours_worked, hourly_rate):
        self.name = name
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate

    def calculate_salary(self):
        return self.hours_worked * self.hourly_rate

class Manager(Employee):
    def __init__(self, name, hours_worked, hourly_rate, bonus):
        super().__init__(name, hours_worked, hourly_rate)
        self.bonus = bonus

    def calculate_salary(self):
        base_salary = super().calculate_salary()
        return base_salary + self.bonus

e = Employee("Alice", 40, 20)
m = Manager("Bob", 40, 25, 500)

print(e.calculate_salary())
print(m.calculate_salary())
```

```python
#15.  Create a class Product with attributes name, price, and quantity. Implement a method total_price() that calculates the total price of the product.
class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def total_price(self):
        return self.price * self.quantity

product = Product("Shirt", 20, 5)
print(product.total_price())

```

```python
#16. Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that implement the sound() method.
from abc import ABC, abstractmethod
class Animal:
    @abstractmethod
    def sound(self):
        pass

class Cow(Animal):
    def sound(self):
        print("Moo!")

class Sheep(Animal):
    def sound(self):
        print("Baa!")

cow = Cow()
sheep = Sheep()

cow.sound()
sheep.sound()
```

```python
#17. Create a class Book with attributes title, author, and year_published. Add a method get_book_info() that returns a formatted string with the book's details.
class Book:
    def __init__(self, title, author, year_published):
        self.title = title
        self.author = author
        self.year_published = year_published

    def get_book_info(self):
        return f"Title: {self.title}\nAuthor: {self.author}\nYear Published: {self.year_published}"

book = Book("The Great Gatsby", "F. Scott Fitzgerald", 1925)
print(book.get_book_info())
```
